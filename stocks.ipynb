```python

ticker_objects = [yf.Ticker(ticker) for ticker in tickers]

data = pd.DataFrame(columns=["Ticker","Gross Profit","ROE","ROA","NDebt/Ebitda","Current Ratio","Interest Cover"])


for ticker_obj in ticker_objects:
    ticker_column=ticker_obj.financials.columns[1:]
    #print(ticker_obj.ticker)
    Gross_Profit = ticker_obj.financials[ticker_column].loc["Gross Profit"]
    Total_Revenue = ticker_obj.financials[ticker_column].loc["Total Revenue"]
    EBITDA = ticker_obj.financials[ticker_column].loc["EBITDA"]
    EBIT = ticker_obj.financials[ticker_column].loc["EBIT"]
    net_profit = ticker_obj.financials[ticker_column].loc["Net Income"]
    equity= ticker_obj.balancesheet[ticker_column].loc["Common Stock Equity"]
    net_debt = ticker_obj.balancesheet[ticker_column].loc["Total Debt"]
    current_assets= ticker_obj.balancesheet[ticker_column].loc["Current Assets"]
    total_assets= ticker_obj.balancesheet[ticker_column].loc["Total Assets"]
    
    current_liabilities =ticker_obj.balancesheet[ticker_column].loc["Current Liabilities"]
    share_issued =ticker_obj.balancesheet[ticker_column].loc["Share Issued"]
    interest_expense= ticker_obj.financials[ticker_column].loc["Interest Expense"]
    operating_cash_flow= ticker_obj.cashflow[ticker_column].loc["Operating Cash Flow"]
    #print(current_liabilities)
    net_profit_current_assets= net_profit/current_assets
    interest_coverage= EBIT/interest_expense
    operating_cash_flow_ratio= operating_cash_flow/current_liabilities
   

    Earnings_shares=EBIT/share_issued
    net_profit_equity_ratio= net_profit/equity
    net_debt_ebitda_ratio= net_debt/EBITDA
    current_assets_current_liabilities_ratio= current_assets/current_liabilities

    gross_profit_ratio = (Gross_Profit / Total_Revenue)
    EBITDA_sales_ratio = EBITDA/Total_Revenue
    EBIT_sales= EBIT/Total_Revenue


    result = pd.DataFrame(gross_profit_ratio, columns=["Gross Profit"])

    result["EBITDA/Trevenue"]=EBITDA_sales_ratio
    result["ROE"]= net_profit_equity_ratio
    result["ROA"]= net_profit_current_assets
    result["Ticker"] = ticker_obj.ticker
    result["NDebt/Ebitda"]= net_debt_ebitda_ratio
    result["Current Ratio"]= current_assets_current_liabilities_ratio
    result["Interest Cover"]= interest_coverage
    result["Earnpershare"]=Earnings_shares
    result["OpCashflow Ratio"]= operating_cash_flow_ratio
    result["EBIT/TRevenue"]=EBIT_sales
    
    
    
   
    data = pd.concat([data, result], ignore_index=True)
data=pd.concat([data,price_increase_percentages_list],axis=1)
print(data)
data.to_csv("financials.csv")
```
